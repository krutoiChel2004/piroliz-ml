```python
import pandas as pd
import numpy as np
```

```python
def generate_pyrolysis_dataset(num_samples=100):
    np.random.seed(42)  # для воспроизводимости
    
    # Генерация входных параметров
    temperature = np.random.uniform(300, 800, num_samples)
    time = np.random.uniform(30, 180, num_samples)
    tyre_type = np.random.choice(['легковая', 'грузовая', 'спецтехника'], num_samples)
    
    # Расчет выходных параметров (упрощенная модель)
    gas_yield = 10 + 0.05 * temperature - 0.001 * time + np.random.normal(0, 5, num_samples)
    liquid_yield = 40 - 0.03 * temperature + 0.002 * time + np.random.normal(0, 5, num_samples)
    carbon_yield = 50 - 0.02 * temperature - 0.001 * time + np.random.normal(0, 5, num_samples)
    
    # Ограничение выхода в пределах 0-100%
    gas_yield = np.clip(gas_yield, 0, 100)
    liquid_yield = np.clip(liquid_yield, 0, 100)
    carbon_yield = np.clip(carbon_yield, 0, 100)
    
    # Создание DataFrame
    data = {
        'Температура (°C)': temperature,
        'Время (мин)': time,
        'Тип_шины': tyre_type,
        'Выход_газа (%)': gas_yield,
        'Выход_жидкости (%)': liquid_yield,
        'Выход_углерода (%)': carbon_yield
    }
    
    df = pd.DataFrame(data)
    return df
```

```python
dataset = generate_pyrolysis_dataset(num_samples=500)  # можно задать любой размер
dataset.head()  # посмотрим первые строки
```

```python
import matplotlib.pyplot as plt

plt.scatter(dataset['Температура (°C)'], dataset['Выход_газа (%)'])
plt.xlabel('Температура (°C)')
plt.ylabel('Выход газа (%)')
plt.title('Зависимость выхода газа от температуры')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import matplotlib.pyplot as plt

# Загрузка данных (или генерация, если файла нет)
try:
    df = pd.read_csv('pyrolysis_dataset.csv')
except:
    # Если файла нет, генерируем заново
    def generate_pyrolysis_dataset(num_samples=500):
        np.random.seed(42)
        temperature = np.random.uniform(300, 800, num_samples)
        time = np.random.uniform(30, 180, num_samples)
        tyre_type = np.random.choice(['легковая', 'грузовая', 'спецтехника'], num_samples)
        
        gas_yield = 10 + 0.05 * temperature - 0.001 * time + np.random.normal(0, 5, num_samples)
        gas_yield = np.clip(gas_yield, 0, 100)
        
        data = {
            'Температура (°C)': temperature,
            'Время (мин)': time,
            'Тип_шины': tyre_type,
            'Выход_газа (%)': gas_yield
        }
        return pd.DataFrame(data)
    
    df = generate_pyrolysis_dataset(500)
    df.to_csv('pyrolysis_dataset.csv', index=False)

# Выведем первые строки
print(df.head())
```

```python
X = df.drop('Выход_газа (%)', axis=1)  # Признаки
y = df['Выход_газа (%)']               # Целевая переменная
```

```python
# Выбираем категориальные колонки
categorical_features = ['Тип_шины']

# Применяем OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_features)
    ],
    remainder='passthrough'  # Остальные колонки оставляем как есть
)

X_processed = preprocessor.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, random_state=42
)

print(f"Обучающая выборка: {X_train.shape}, Тестовая выборка: {X_test.shape}")
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Предсказание на тестовых данных
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Среднеквадратичная ошибка (MSE): {mse:.2f}")
print(f"Коэффициент детерминации (R²): {r2:.2f}")
```

```python
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # Идеальная линия
plt.xlabel("Истинные значения")
plt.ylabel("Предсказанные значения")
plt.title("Предсказание выхода газа (%)")
plt.show()
```
